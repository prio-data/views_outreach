## Fetch and transform remaining features
The three features below could not be automatically aggregated to country_year level due to their countrylag transforms and are therefore processed separately. Here, we fetch them at the original country_month level, group them by year, and create a mean of the values per year. The result is found the Batch2 csv file. 

**Note:** Only the original country-month versions of these features are defined in the prediction challenge codebook.

```python
## Set time period: 1980-2020

StartMonth = 1 # month_id 1 is 1980-1. NOTE: There seems to be a lot of missing data in 1980-1981. 
EndMonth = 492 
```

### Fetch the data

```python
predcomp_subset_countrylag_features = (Queryset('predcomp_subset_countrylag_features','country_month')
                             
    .with_column(Column('year', from_loa = "year", from_column = 'year'))
    .with_column(Column('month', from_loa = "month", from_column = 'month'))
    .with_column(Column('isoab', from_loa = "country", from_column = 'isoab'))
    .with_column(Column('country', from_loa = "country", from_column = 'name'))

    .with_column(Column('splag_1_decay_ged_sb_5', from_loa='country_month', from_column='ged_sb_best_sum_nokgi')
        .transform.missing.replace_na()
        .transform.bool.gte(5)
        .transform.temporal.time_since()
        .transform.temporal.decay(24)
        .transform.spatial.countrylag(1,1,0,0)
        .transform.missing.replace_na()
        )

    .with_column(Column('splag_1_decay_ged_os_5', from_loa='country_month', from_column='ged_os_best_sum_nokgi')
        .transform.missing.replace_na()
        .transform.bool.gte(5)
        .transform.temporal.time_since()
        .transform.temporal.decay(24)
        .transform.spatial.countrylag(1,1,0,0)
        .transform.missing.replace_na()
        )

    .with_column(Column('splag_1_decay_ged_ns_5', from_loa='country_month', from_column='ged_ns_best_sum_nokgi')
        .transform.missing.replace_na()
        .transform.bool.gte(5)
        .transform.temporal.time_since()
        .transform.temporal.decay(24)
        .transform.spatial.countrylag(1,1,0,0)
        .transform.missing.replace_na()
        )
)

predcomp_subset_countrylag_features = predcomp_subset_countrylag_features.publish().fetch()

print(f"A dataset with {len(predcomp_subset_countrylag_features.columns)} columns, with "
      f"data between t {min(predcomp_subset_countrylag_features.index.get_level_values(0))} "
      f"and {max(predcomp_subset_countrylag_features.index.get_level_values(0))}. "
      f"({len(np.unique(predcomp_subset_countrylag_features.index.get_level_values(1)))} units)"
     )
```

```python
# Reset the index to get a clean dataframe

predcomp_subset_countrylag_features_filtered = predcomp_subset_countrylag_features_filtered.reset_index()
predcomp_subset_countrylag_features_filtered
```

```python
# Save the pre-processed to a CSV file

predcomp_subset_countrylag_features_filtered.to_csv(desktop+'predcomp_subset_countrylag_features_pre-processed.csv', index = False)
```

### Aggregate the data and calculate the mean per year

```python
# Group by country_id, country, isoab, year and then take the mean of the numeric columns over each year 

mean_per_year = predcomp_subset_countrylag_features_filtered.groupby(["country_id", "country", "isoab", "year"]).mean(numeric_only=True).drop(columns=["month_id", "month"])
mean_per_year
```
